# Train and host model using Amazon Lookout for Vision and predict using AWS CLI

## ** BEGIN **  
## Data Preparation:- Please follow below instructions.

1. Download the labelled dataset from Kaggle.com. => https://www.kaggle.com/anujms/car-damage-detection
2. Create a folder locally "Lookout4Vision" with a subfolder "training" and 2 subfolders underneath as "Anomaly" & "Normal". 
3. Copy images from "training/00-damage" => "training/Anomaly" & "training/01-whole" => "training/Normal".
4. Copy images from "validation" dataset as well. These would be used later to run trails & predictions.

Note:- Lookout Vision supports Image dimension between 64 pixels x 64 pixels (min.) to 4096 pixels X 4096 pixels (max.) and  PNG & JPEG image formats. All images in the dataset should have same dimensions.

## Setup Lookout For Vision: Create s3 bucket, train model & run trails:

5.	Select an AWS region, where Lookout Vision is available. 
Lookout for Vision is available today in US East ( N. Virginia ), US East ( Ohio ), US West ( Oregon ), EU ( Ireland ), EU ( Frankfort ), Asia Pacific ( Tokyo ), and Asia Pacific ( Seoul ), with availability in additional regions in the coming months.

6.	Create an Amazon S3 bucket in same region as Lookout Vision & upload the labelled data in same order.

7.	Go to Amazon Lookout for Vision => Create Project =>


8.	Create Dataset, replace “xyz” with your s3 bucket name & select automatically attach label.



9.	Click on Train Model.




10.	Run Trail detection to verify model predictions on new data sets (Use same dimensions of images as used in training data, Use trail or validation data set images, refer step-4).



### Note:- Please install aws cli for performing below steps.
https://docs.aws.amazon.com/cli/latest/userguide/install-cliv2.html

12.	 Host Model  (Model status would change from “Training Complete” to “Hosted”) 
 Use below snippets: <<Use your project name in place of "Detect-Defect-Car-2">> refer step-7.

aws lookoutvision start-model \
  --project-name Detect-Defect-Car-2 \
  --model-version 1 \
  --min-inference-units 1

###### Output:- Visualize output using aws console.




## Make Predictions using AWS CLI

Use below snippets:

aws lookoutvision detect-anomalies \
  --project-name Detect-Defect-Car-2 \
  --model-version 1 \
  --content-type image/jpeg \
  --body /path/to/image.jpeg
